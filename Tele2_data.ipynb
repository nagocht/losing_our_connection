```python
import json
import pandas as pd
import codecs
from pprint import pprint

#Exploring the data from TELE2

data = []

#The following features were interesting to our project
columnlist = ['timestamp', 'lat', 'lon','locationAccuracy','accountName','domainId', 'id']

with open('vasttrafik.json') as file:
    for line in file:
        #Pre-processing the data
        parts = line.split(' ',1)
        parts[1] = json.loads(parts[1])
        #There were 2 kinds of data, one called (2control/synthesized_events) and one called (2control/protocol)
        #We were only interested in the control data so we filter based on our desired columns
        try:
            timestamp = parts[1]['timestamp']
            lat = parts[1]['payload']['derived']['location']['lat']
            lon = parts[1]['payload']['derived']['location']['lon']
            locationAccuracy = parts[1]['payload']['derived']['locationAccuracy']
            accountName = parts[1]['payload']['derived']['subscriber']['accountName']
            domainID = parts[1]['payload']['derived']['subscriber']['domainId']
            ID = parts[1]['payload']['derived']['subscriber']['id']
            data.append([timestamp, lat, lon, locationAccuracy, accountName, domainID, ID])
        except:
            pass
df = pd.DataFrame(data, columns = columnlist)
df.head()
```
